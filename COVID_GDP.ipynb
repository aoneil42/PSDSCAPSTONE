```r
COVID <- read.csv("COVID_Totals_Demo.csv")
GDP_Q2 <- read.csv("GDP_2Q.csv")
GDP_Q3 <- read.csv("GDP_3Q.csv")
```

```r
COVID$Country <- as.character(COVID$Country)
COVID[COVID$Country == "US",]$Country <- c("United States")
COVID[COVID$Country == "China",]$Country <- c("China (People's Republic of)")
COVID[COVID$Country == "Korea, South",]$Country <- c("Korea")
COVID[COVID$Country == "Czechia",]$Country <- c("Czech Republic")
```

```r
COV_GDP_Q2 <- merge(COVID,GDP_Q2,by="Country")[,-c(2,18)]
COV_GDP_Q3 <- merge(COVID,GDP_Q3,by="Country")[,-c(2,18)]
```

```r
rownames(COV_GDP_Q2) <- COV_GDP_Q2$Country
COV_GDP_Q2.sub <- COV_GDP_Q2[,-c(1,4,7,10,13,17,18,19,20)]
head(COV_GDP_Q2.sub)
```

```r
rownames(COV_GDP_Q3) <- COV_GDP_Q3$Country
COV_GDP_Q3.sub <- COV_GDP_Q3[,-c(1,17,18,19,20)]
head(COV_GDP_Q3.sub[,c(6,12,13,14,15,18)])
```

```r
Q3case.sub <- COV_GDP_Q3.sub[,c(6,13,14,15,18)]
Q3death.sub <- COV_GDP_Q3.sub[,c(12,13,14,15,18)]
cor(Q3case.sub)
cor(Q3death.sub)
```

```r
Q3case.sub.lm <- lm(Q3rate100kcases ~ ., data=Q3case.sub)
Q3death.sub.lm <- lm(Q3rate100kdeaths ~ ., data=Q3death.sub)
Q3case.sub.lmw <- lm(Q3rate100kcases ~ ., data=Q3case.sub, weights=Density2018)
Q3death.sub.lmw <- lm(Q3rate100kdeaths ~ ., data=Q3death.sub, weights=Density2018)
```

```r
plot(Q3case.sub.lmw)
```

```r
plot(Q3death.sub.lm)
```

```r
Q2death.sub <- COV_GDP_Q2.sub[,c(8,9,10,11,13)]
Q2case.sub <- COV_GDP_Q2.sub[,c(4,9,10,11,13)]
```

```r
Q2case.sub.lmgdp <- lm(Q2_2020 ~ Q2rate100kcases, data=Q2case.sub)
Q2death.sub.lmgdp <- lm(Q2_2020 ~ Q2rate100kdeaths, data=Q2death.sub)
Q2case.sub.lmgdpw <- lm(Q2_2020 ~ Q2rate100kcases, data=Q2case.sub, weights=Density2018)
Q2death.sub.lmgdpw <- lm(Q2_2020 ~ Q2rate100kdeaths, data=Q2death.sub, weights=Density2018)
Q3case.sub.lmgdp <- lm(Q3_2020 ~ Q3rate100kcases, data=Q3case.sub)
Q3death.sub.lmgdp <- lm(Q3_2020 ~ Q3rate100kdeaths, data=Q3death.sub)
Q3case.sub.lmgdpw <- lm(Q3_2020 ~ Q3rate100kcases, data=Q3case.sub, weights=Density2018)
Q3death.sub.lmgdpw <- lm(Q3_2020 ~ Q3rate100kdeaths, data=Q3death.sub, weights=Density2018)
```

```r
plot(Q3case.sub.lmgdpw)
```

```r
Q2case.sub.lmgdp.out <- c("China (People's Republic of)","India","United Kingdom", "Spain")
Q2case.sub.lmgdpw.out <- c("China (People's Republic of)","India", "Korea","United Kingdom", "Luxembourg")
Q3case.sub.lmgdp.out <- c("China (People's Republic of)","Israel","United Kingdom", "Chile")
Q3case.sub.lmgdpw.out <- c("China (People's Republic of)","Israel", "United Kingdom", "Chile", "Korea")

Q2death.sub.lmgdp.out <- c("China (People's Republic of)","India","Argentina","Belgium", "Spain")
Q2death.sub.lmgdpw.out <- c("China (People's Republic of)","India","Argentina", "Korea",
                            "Belgium", "United Kingdom")
Q3death.sub.lmgdp.out <- c("China (People's Republic of)","Chile","Colombia","United States","Belgium")
Q3death.sub.lmgdpw.out <- c("China (People's Republic of)","Chile","Colombia", "United Kingdom"
                            , "Japan","Belgium")

```

```r
Q2case.sub2 <- Q2case.sub[rownames(Q2case.sub) %in% c(Q2case.sub.lmgdp.out,
                                       Q2case.sub.lmgdpw.out),]
Q2case.sub2$Country <- rownames(Q2case.sub2)
Q3case.sub2 <- Q3case.sub[rownames(Q3case.sub) %in% c(Q3case.sub.lmgdp.out,
                                       Q3case.sub.lmgdpw.out),]
Q3case.sub2$Country <- rownames(Q3case.sub2)
Q2death.sub2 <- Q2death.sub[rownames(Q2death.sub) %in% c(Q2death.sub.lmgdp.out,
                                       Q2death.sub.lmgdpw.out),]
Q2death.sub2$Country <- rownames(Q2death.sub2)
Q3death.sub2 <- Q3death.sub[rownames(Q3death.sub) %in% c(Q3death.sub.lmgdp.out,
                                       Q3death.sub.lmgdpw.out),]
Q3death.sub2$Country <- rownames(Q3death.sub2)
```

```r
countries.selected <- unique(Reduce(function(...) merge(...,all=T),
              (list(Q2case.sub2,Q3case.sub2,Q2death.sub2,Q3death.sub2)))[,c(1,4)])
```

```r
write.csv(countries.selected, "countries.selected.csv")
```
